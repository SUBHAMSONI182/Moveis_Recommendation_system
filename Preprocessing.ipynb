```python
def convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
import ast
```

```python
moveis['genres']=moveis['genres'].apply(convert)
```

```python
moveis['keywords']=moveis['keywords'].apply(convert)
```

```python
moveis.head()
```

```python
def convert3(obj):
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
        
    return L
```

```python
moveis['cast']= moveis['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L =[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
moveis['crew']=moveis['crew'].apply(fetch_director)
```

```python
moveis['overview']=moveis['overview'].apply(lambda x:x.split())
```

```python
moveis.head()
```

```python
moveis['genres']=moveis['genres'].apply(lambda x:[i.replace(" ","") for i in x])
moveis['keywords']=moveis['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
moveis['crew']=moveis['crew'].apply(lambda x:[i.replace(" ","") for i in x])
moveis['cast']=moveis['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
moveis.head()
```

```python
moveis['tags']=moveis['overview']+moveis['genres']+moveis['keywords']+moveis['crew']+moveis['cast']
```

```python
moveis.head()
```

```python
new_df = moveis[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
import  nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))


    return " ".join(y)
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
#text vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
cv.get_feature_names_out()
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
list(enumerate(similarity[0]))
```

```python
new_df[new_df['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
